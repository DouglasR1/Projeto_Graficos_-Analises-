```python
# Importando as Bibliotecas

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criando variável X com números aleatórios 

X = np.random.rand(2000)
```

```python
# Criando Histograma (Gráfico)

plt.hist(X)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.show()
```

```python
# Alterando número de bins

plt.hist(X,bins=10)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=10')
plt.show()
```

```python
# Alterando a largura do Gráfico para uma melhor Visualização

plt.hist(X,bins=20)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=20')
plt.show()
```

```python
plt.hist(X,bins=30)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=30')
plt.show()
```

```python
# Alterando as Cores 

plt.hist(X,bins=50,color='magenta')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
plt.hist(X,bins=50,color='beige')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
plt.hist(X,bins=50,color='turquoise')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
# Criando as entrelinhas das barras do Histograma

plt.hist(X,bins=50,color='turquoise', edgecolor='black')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
# Multiplos Histogramas na mesma Imagem

X1 = np.random.normal(0,1,size = 800)
X2 = np.random.normal(3,1,size = 800)
X3 = np.random.normal(5,1,size = 800)


     

# Criando o Gráfico

plt.hist(X1,bins=10,color='violet')
plt.hist(X2,bins=10,color='magenta')
plt.hist(X3,bins=10,color='purple')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Multiplos histogramas')
plt.show()
```

```python
# Distribuição cumulativa

X = np.random.random(2000)
     

# Criando o Gráfico

plt.hist(X,bins=20,color='turquoise', edgecolor='black',cumulative=True)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```
